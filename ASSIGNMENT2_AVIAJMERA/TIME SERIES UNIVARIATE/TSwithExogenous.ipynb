```python
from pycaret.time_series import *
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset.set_index('Date', inplace=True)


```

```python
# Specify the target variable
a= setup(data=dataset, target='meantemp', fh=3, session_id=123)
b= setup(data=dataset, target='humidity', fh=3, session_id=123)
c= setup(data=dataset, target='wind_speed', fh=3, session_id=123)
d=setup(data=dataset, target='meanpressure', fh=3, session_id=123)
```

```python
check_stats()
```

```python
best = compare_models()
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 50})
```

```python
plot_model(best, plot = 'residuals')
```

```python
holdout_pred = predict_model(best)
```
