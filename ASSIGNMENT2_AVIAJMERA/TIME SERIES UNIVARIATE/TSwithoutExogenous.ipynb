## Importing dataset

```python
dataset = pd.read_csv('/content/TS006-daily-total-female-births.csv')
dataset
```

```python
dataset.plot(x='Date',y='Births')
```

```python
from pycaret.time_series import *
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset.set_index('Date', inplace=True)

# Specify the target variable
s = setup(data=dataset, target='Births', fh=3, session_id=123)
```

```python
check_stats()
```

```python
best = compare_models()
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 50})
```

```python
plot_model(best, plot = 'residuals')
```

```python
holdout_pred = predict_model(best)
```

```python
predict_model(best, fh =50)
```
